```python
words = ['apple', 'bat', 'bar', 'atom', 'book']
```

```python
by_letter = dict()
```

```python
for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
```

```python
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
```

```python
states = [' Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda','south carolina##', 'West virginia?']
```

```python
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = re.sub('[!#?]','',value)
        value = value.title()
        result.append(value)
    return result
```

```python
def single_clean_strings(value):
    # result = []
    # for value in strings:
    value = value.strip()
    value = re.sub('[!#?]','',value)
    value = value.title()
    # result.append(value)
    return value
```
